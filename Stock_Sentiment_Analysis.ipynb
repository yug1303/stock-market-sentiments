### Train Test Splitting

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

### Data Preprocessing

```python
#Removing punctations
data = df.iloc[:,2:27]
data.replace("[^a-zA-Z]", " ",regex=True, inplace=True)
data.replace("['b]", "",regex=True, inplace=True)


#Reaming ames for columns for ease of acces
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```python
#Converting headlines to lower case
for index in new_index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[0,:25])
```

```python
headlines = []
for row in range(0,len(train.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

### Feature extraction from text and model selection

```python
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#implememt BAG OF WORDS
countVector = CountVectorizer(ngram_range=(2,2))
#ngram(2,2) means it will combine the 2 words together and assign the value

trainDataset = countVector.fit_transform(headlines)
```

```python
trainDataset[0]
```

```python
#implement random forest classifier
randomclassifier = RandomForestClassifier(criterion='entropy', n_estimators=200)
randomclassifier.fit(trainDataset, train['Label'])
```

### Predict for the Test Dataset

```python
testTransform =[]
for row in range(0, len(test.index)):
    testTransform.append(' '.join(str(x) for x in test.iloc[row, 2:27]))
    
test_dataset = countVector.transform(testTransform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

### Check accuracy

```python
#importing libraries to check accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```
