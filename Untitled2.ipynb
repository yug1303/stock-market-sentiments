Step 2: Data Loading and Preparation

```python
# Load the dataset
df = pd.read_csv('Data.csv', encoding='ISO-8859-1')

# Split the dataset into train and test sets
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']

# Removing punctuations and making text lowercase
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
data.columns = [str(i) for i in range(25)]

# Convert headlines to lowercase
for index in data.columns:
    data[index] = data[index].str.lower()

# Combine all the headlines into a single string for each row
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))

# Vectorizing the combined headlines using CountVectorizer
countvector = CountVectorizer(ngram_range=(2, 2))
traindataset = countvector.fit_transform(headlines)

# Train the RandomForest Classifier
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindataset, train['Label'])
```

Step 3: Making Predictions and Evaluating the Model

```python
# Prepare the test dataset
test_transform = []
for row in range(0, len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row, 2:27]))

test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)

# Evaluate the model
matrix = confusion_matrix(test['Label'], predictions)
score = accuracy_score(test['Label'], predictions)
report = classification_report(test['Label'], predictions)

print("Confusion Matrix:\n", matrix)
print("Accuracy Score:", score)
print("Classification Report:\n", report)
```

Step 6: Calculate Portfolio Returns and Performance Metrics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic stock data
dates = pd.date_range(start='2014-01-01', periods=500, freq='D')
price_changes = np.random.normal(loc=0.0005, scale=0.01, size=(500,))
prices = 100 * (1 + price_changes).cumprod()

stock_data = pd.DataFrame(data={'Close': prices}, index=dates)

# Initialize variables
initial_balance = 100000
balance = initial_balance
positions = 0
portfolio_value = []

# Calculate the moving average
short_window = 4
long_window = 100
stock_data['short_ma'] = stock_data['Close'].rolling(window=short_window).mean()
stock_data['long_ma'] = stock_data['Close'].rolling(window=long_window).mean()

# Buy when the short MA crosses above the long MA
# Sell when the short MA crosses below the long MA
for i in range(len(stock_data)):
    if stock_data['short_ma'].iloc[i] > stock_data['long_ma'].iloc[i] and positions == 0:
        positions = balance / stock_data['Close'].iloc[i]
        balance = 0
    elif stock_data['short_ma'].iloc[i] < stock_data['long_ma'].iloc[i] and positions > 0:
        balance = positions * stock_data['Close'].iloc[i]
        positions = 0
    portfolio_value.append(balance + positions * stock_data['Close'].iloc[i])

# Convert portfolio value to a DataFrame
portfolio_value = pd.DataFrame(portfolio_value, index=stock_data.index, columns=['Portfolio Value'])

# Calculate returns and performance metrics
returns = portfolio_value['Portfolio Value'].pct_change().dropna()
sharpe_ratio = (returns.mean() / returns.std()) * np.sqrt(252)  # Annualized Sharpe Ratio

# Calculate maximum drawdown
rolling_max = portfolio_value['Portfolio Value'].cummax()
drawdown = portfolio_value['Portfolio Value'] / rolling_max - 1
max_drawdown = drawdown.min()

# Calculate number of trades and win ratio
trade_signals = ((stock_data['short_ma'] > stock_data['long_ma']) & (stock_data['short_ma'].shift(1) <= stock_data['long_ma'].shift(1))) | \
                ((stock_data['short_ma'] < stock_data['long_ma']) & (stock_data['short_ma'].shift(1) >= stock_data['long_ma'].shift(1)))
trades = trade_signals.sum()

buy_signals = (stock_data['short_ma'] > stock_data['long_ma']) & (stock_data['short_ma'].shift(1) <= stock_data['long_ma'].shift(1))
sell_signals = (stock_data['short_ma'] < stock_data['long_ma']) & (stock_data['short_ma'].shift(1) >= stock_data['long_ma'].shift(1))

win_trades = sum(stock_data['Close'][sell_signals].values > stock_data['Close'][buy_signals].values[:len(stock_data['Close'][sell_signals])])
win_ratio = win_trades / (trades / 2)

# Plot portfolio value
plt.figure(figsize=(14, 7))
plt.plot(stock_data['Close'], label='Stock Price')
plt.plot(stock_data['short_ma'], label=f'Short MA ({short_window} days)')
plt.plot(stock_data['long_ma'], label=f'Long MA ({long_window} days)')
plt.plot(portfolio_value, label='Portfolio Value', color='black')
plt.legend(loc='best')
plt.title('Stock Price and Portfolio Value with Moving Average Crossover Strategy')
plt.show()

# Print performance metrics
print("Sharpe Ratio:", sharpe_ratio)
print("Maximum Drawdown:", max_drawdown)
print("Number of Trades Executed:", trades)
print("Win Ratio:", win_ratio)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample stock data generation (replace this with your actual stock data)
dates = pd.date_range(start='2020-01-01', end='2023-01-01', freq='B')
prices = np.cumsum(np.random.randn(len(dates)) * 2 + 0.1) + 100  # Simulated stock prices
stock_data = pd.DataFrame(data={'Close': prices}, index=dates)

# Initialize variables for portfolio calculation
initial_balance = 100000
balance = initial_balance
positions = 0
portfolio_value = []

# Calculate the moving average
short_window = 4
long_window = 100
stock_data['short_ma'] = stock_data['Close'].rolling(window=short_window).mean()
stock_data['long_ma'] = stock_data['Close'].rolling(window=long_window).mean()

buy_signals = []
sell_signals = []

# Moving Average Crossover Strategy
for i in range(len(stock_data)):
    if stock_data['short_ma'].iloc[i] > stock_data['long_ma'].iloc[i] and positions == 0:
        positions = balance / stock_data['Close'].iloc[i]
        balance = 0
        buy_signals.append(stock_data.index[i])
    elif stock_data['short_ma'].iloc[i] < stock_data['long_ma'].iloc[i] and positions > 0:
        balance = positions * stock_data['Close'].iloc[i]
        positions = 0
        sell_signals.append(stock_data.index[i])
    portfolio_value.append(balance + positions * stock_data['Close'].iloc[i])

# Convert portfolio value to a DataFrame
portfolio_value = pd.DataFrame(portfolio_value, index=stock_data.index, columns=['Portfolio Value'])

# Calculate returns and performance metrics for moving average strategy
returns = portfolio_value['Portfolio Value'].pct_change().dropna()
sharpe_ratio = (returns.mean() / returns.std()) * np.sqrt(252)  # Annualized Sharpe Ratio

# Calculate maximum drawdown for moving average strategy
rolling_max = portfolio_value['Portfolio Value'].cummax()
drawdown = portfolio_value['Portfolio Value'] / rolling_max - 1
max_drawdown = drawdown.min()

# Calculate number of trades and win ratio for moving average strategy
trades = 0
wins = 0
for i in range(1, len(stock_data)):
    if stock_data['short_ma'].iloc[i] > stock_data['long_ma'].iloc[i-1] and stock_data['short_ma'].iloc[i-1] <= stock_data['long_ma'].iloc[i-1]:
        trades += 1  # Buy signal
    elif stock_data['short_ma'].iloc[i] < stock_data['long_ma'].iloc[i-1] and stock_data['short_ma'].iloc[i-1] >= stock_data['long_ma'].iloc[i-1]:
        trades += 1  # Sell signal
        if stock_data['Close'].iloc[i] > stock_data['Close'].iloc[i-1]:
            wins += 1

win_ratio = wins / (trades / 2) if trades > 0 else 0

# Buy and Hold Strategy
buy_and_hold_balance = initial_balance / stock_data['Close'].iloc[0] * stock_data['Close'].iloc[-1]
buy_and_hold_return = (buy_and_hold_balance - initial_balance) / initial_balance
buy_and_hold_annualized_return = ((1 + buy_and_hold_return) ** (252 / len(stock_data)) - 1)
buy_and_hold_volatility = stock_data['Close'].pct_change().std() * np.sqrt(252)
buy_and_hold_sharpe_ratio = buy_and_hold_annualized_return / buy_and_hold_volatility

# Print results
print("Moving Average Strategy:")
print("Sharpe Ratio:", sharpe_ratio)
print("Maximum Drawdown:", max_drawdown)
print("Number of Trades Executed:", trades)
print("Win Ratio:", win_ratio)

# print("\nBuy and Hold Strategy:")
# print("Annualized Return:", buy_and_hold_annualized_return)
# print("Sharpe Ratio:", buy_and_hold_sharpe_ratio)

# Plotting the results
plt.figure(figsize=(14, 7))
plt.plot(stock_data['Close'], label='Stock Price', alpha=0.5)
plt.plot(stock_data['short_ma'], label=f'Short {short_window}-Day MA', alpha=0.7)
plt.plot(stock_data['long_ma'], label=f'Long {long_window}-Day MA', alpha=0.7)
plt.scatter(buy_signals, stock_data.loc[buy_signals]['Close'], marker='^', color='green', label='Buy Signal', alpha=1)
plt.scatter(sell_signals, stock_data.loc[sell_signals]['Close'], marker='v', color='red', label='Sell Signal', alpha=1)
plt.title('Stock Price and Moving Average Crossover Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(portfolio_value, label='Portfolio Value - Moving Average Strategy', alpha=0.7)
plt.title('Portfolio Value Over Time')
plt.xlabel('Date')
plt.ylabel('Portfolio Value')
plt.legend()
plt.show()
```
